**Importo las librerías necesarias**

```python
import requests
import csv
import pandas as pd
```

**Realizo una solicitud a una API de Criptomonedas llamada *CoinCap 2.0***

*CoinCap es una herramienta útil para la fijación de precios en tiempo real y la actividad del mercado para más de 1.000 criptomonedas. Al recopilar datos bursátiles de miles de mercados, podemos ofrecer datos transparentes y precisos sobre el precio y la disponibilidad de los activos.*

```python
response = requests.get('https://api.coincap.io/v2/assets')
response
```

*La conexión fue exitosa*

**Convierto la respuesta JSON en una lista de diccionarios**

```python
data = response.json()
data
```

**De la lista de diccionarios elimino 'timestamp' que no lo necesito para trabajar el dataset**

```python
del data['timestamp']
data
```

**Luego accedo a los datos agrupados en 'data' con el siguiente código:**

```python
data2 = data['data']
data2
```

**Convierto los datos en un Dataframe**

```python
df = pd.DataFrame(data2)
df
```
